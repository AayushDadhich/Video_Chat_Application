```python
# import the opencv librar
%matplotlib
import cv2
import socket
from pickle import dumps,loads
import matplotlib.pyplot as plt


# create an INET, STREAMing socket
csocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
csocket.bind(("localhost",5000))
# bind the socket to a public host, and a well-known port
csocket.listen(1)
server, caddr = csocket.accept()
#server.connect(("localhost",5001))

ax = plt.subplot(1,1,1)
temp = ax.imshow([[1]])
plt.pause(1)
plt.ion()

# become a server socket
#server.listen(1)

# define a video capture object 
#csocket, caddr = server.accept()
#print("Connected.")

while(True): 
    vid = cv2.VideoCapture(0)
    #print(request)
    ret, frame = vid.read() 
    #print(frame)
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    #print(len(dumps(img)))
    server.send(dumps(img))
    vid.release()
    #csocket.send(dumps(None))
    #print("Sent")
    data = server.recv(921764)
    #print(len(data))
    #print("Image Received")
    img = loads(data)
    #print(type(img))
    temp.set_data(img)
    plt.pause(0.01)
    #print(type(img))
    
server.close()
csocket.close() # state less protocol
# After the loop release the cap object 

# Destroy all the windows 
cv2.destroyAllWindows() 
```
