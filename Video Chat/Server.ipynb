```python
# import the opencv librar
import cv2
import socket
from pickle import dumps


# create an INET, STREAMing socket
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# bind the socket to a public host, and a well-known port
server.bind(('localhost', 5001))

# become a server socket
server.listen(1)

# define a video capture object 
 
c=0
print("__"*30)
print(f"Client number.")
csocket, caddr = server.accept()
print("Got Connected to a client : ", caddr)
vid = cv2.VideoCapture(0)
while(True): 
    request = csocket.recv(1024).decode()
    #print(request)
    ret, frame = vid.read() 
    #print(frame)
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    #print(len(dumps(img)))
    csocket.send(dumps(img))
    #csocket.send(dumps(None))
    #print("Sent")
     
csocket.close() # state less protocol
# After the loop release the cap object 
vid.release()
# Destroy all the windows 
cv2.destroyAllWindows() 
```
